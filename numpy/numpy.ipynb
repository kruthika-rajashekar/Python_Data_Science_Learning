Small game using randint

```python
import numpy as np

n=np.random.randint(1,100)
count=0
while True:
    user=int(input("Enter a guessed number"))
    count+=1 
    if user==n:
        print("You have entered correct random number")
        break
    elif user<n:
        print("Random number is greater than what you entered")
    elif user>n:
        print("Random number is less than what you entered")
print(count)
```

```python
l=['red','blue','green']

np.random.choice(l)
```

```python
choice=['stone','paper','scissor']
u_count=0
n_count=0
for i in range(1,4):
    n=np.random.choice(choice)
    print(f"Computer Choice: {n}")
    user=input("Enter your choice : ")
    print(f"User Choice: {user}")
    if user==n:
        pass
    elif (user=='stone' and n=='paper') or (user=='paper' and n=='scissor') or (user=='scissor' and n=='stone'):
        n_count+=1
    else:
        u_count+=1

print(n_count)
print(u_count)

if n_count > u_count:
    print("computer won the game")
elif n_count==u_count:
    print("Both are tie")
else:
    print("User won the game")    
```

```python
#Variance of X

x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])

mean_x=np.mean(x)
n=0
for i in x:
    sum=0
    n+=1
    b=(i-mean_x)
    sum+=b
    sum_2=sum**2
var_x=sum_2/n

var_x
```

Variance of x

```python
var_x=np.sum((x-np.mean(x))**2)/n
var_x
```

Variance of Y

```python
var_y=np.sum((y-np.mean(y))**2)/n
var_y
```

Covariance of (X,Y)

```python
Cov_XY=np.sum((x-np.mean(x))*(y-np.mean(y)))/n
Cov_XY
```
