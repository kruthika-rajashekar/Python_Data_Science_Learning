```python
ds.fillna({'Postal Code':0})
```

```python
ds.dropna(subset=['Sales'])
```

```python
ds.fillna({'Profit':ds.agg({'Profit':'mean'})})
```

```python
total=ds.groupby('Region')
total.agg({'Sales':'sum'})
```

```python
total=ds.groupby('Region').agg({'Sales':'sum'})
total
```

```python
avg=ds.groupby('Category').agg({'Sales':'mean'})
avg
```

```python
Maximum_Profit=ds.groupby('Category').agg({'Profit':'max'})
Maximum_Profit
```

```python
Minimum_Profit=ds.groupby('Category').agg({'Profit':'min'})
Minimum_Profit
```

```python
Total_Quantity=ds.groupby('Category').agg({'Quantity':'sum'})
Total_Quantity
```

```python
sum_mean=ds.groupby('Region').agg({'Profit':['sum','mean']})
sum_mean
```

```python
mean_max=ds.groupby('Category').agg({'Sales':['mean','max']})
mean_max
```
