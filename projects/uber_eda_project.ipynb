Year wise average fare amounts

Weekday average fare amounts

finded using aggregation function


```python
df.groupby('key_weekday')['fare_amount'].mean().sort_values()
```

Conclusion :

Least average fare amount on the weekday '5' is '11.032273'

Most average fare amount on the weekday '6' is '11.756463'

So, if we increase the number of riders on weekday 6 we'll get more profit


```python
df.groupby('key_weekday')['fare_amount'].sum().sort_values()
```
